# Bonus Analysis: Top Products by Revenue

This bonus analysis identifies the top-selling products based on total sales revenue across the assigned territories.

```python
# Import pandas for data analysis
import pandas as pd

# Import matplotlib for chart creation
import matplotlib.pyplot as plt
```

```python
# Load the store sales dataset
store_sales = pd.read_csv('StoreSales.csv')

# Load the products dataset
products = pd.read_csv('Products.csv')

# Load the store detail dataset
store_detail = pd.read_csv('StoreDetail.csv')
```

```python
# Merge sales data with product data using product number
sales_products = pd.merge(
    store_sales,
    products,
    on='Prod Num'
)

# Preview merged dataset
sales_products.head()

# This merge connects each sale transaction to a product name.
```

```python
# Group by product name and calculate total sales revenue
top_products = sales_products.groupby(
    'Product'
)['Sale Amount'].sum().sort_values(
    ascending=False
).head(10)

# Display top 10 products by revenue
top_products

# Sorting highest-to-lowest helps identify the strongest-selling products.
```

```python
# Create figure size for better readability
plt.figure(figsize=(12,5))

# Create bar chart
top_products.plot(
    kind='bar'
)

# Add chart title
plt.title('Top 10 Products by Revenue')

# Add x-axis label
plt.xlabel('Product')

# Add y-axis label
plt.ylabel('Total Revenue')

# Rotate product names for readability
plt.xticks(rotation=45)

# Adjust spacing to prevent label cutoff
plt.tight_layout()

# Display chart
plt.show()
```

The chart above highlights the top-performing individual products based on total sales revenue.

Products generating the highest revenue may represent strong customer demand and profitable inventory opportunities.

This analysis could help marketing teams prioritize promotions, product placement, and inventory planning for future quarters.
